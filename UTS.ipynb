### 3
Create a function that return employees with it's company domain.

Expected return:


```json
[{
    "full_name": "John Doe",
    "company": "Company 1",
    "domain": "Retail"
}, {
    "full_name": "Tom Smith",
    "company": "Company 2",
    "domain": "Construction"
}, {
    "full_name": "Andrew Sebastian",
    "company": "Company 2",
    "domain": "Construction"
}]
```

````python
def get_employees_with_domain(employees, companies):
    employees_with_domain = []
    for employee in employees:
        for company in companies:
            if employee['company'] == company['name']:
                employee_data = {
                    "full_name": employee['full_name'],
                    "company": employee['company'],
                    "domain": company['domain']
                }
                employees_with_domain.append(employee_data)
                break
    return employees_with_domain

# Data karyawan
employees = [
    {"full_name": "John Doe", "company": "Company 1"},
    {"full_name": "Tom Smith", "company": "Company 2"},
    {"full_name": "Andrew Sebastian", "company": "Company 2"}
]

# Data perusahaan
companies = [
    {"name": "Company 1", "domain": "Retail"},
    {"name": "Company 2", "domain": "Construction"}
]

# Memanggil fungsi dan mencetak hasilnya
result = get_employees_with_domain(employees, companies)
print(result)
````

### 4
Create a function that return companies with list of employees that are working in the company.

Expected return:


```json
[{
    "company": "Company 1",
    "employees": ["John Doe"]
}, {
    "company": "Company 2",
    "employees": ["Tom Smith", "Andrew Sebastian"]
}, {
    "company": "Company 3",
    "employees": []
}]
```

````python
def get_companies_with_employees(employees, companies):
    companies_with_employees = []
    for company in companies:
        employees_in_company = [employee['full_name'] for employee in employees if employee['company'] == company['name']]
        companies_with_employees.append({"company": company['name'], "employees": employees_in_company})
    return companies_with_employees

# Data karyawan
employees = [
    {"full_name": "John Doe", "company": "Company 1"},
    {"full_name": "Tom Smith", "company": "Company 2"},
    {"full_name": "Andrew Sebastian", "company": "Company 2"}
]

# Data perusahaan
companies = [
    {"name": "Company 1"},
    {"name": "Company 2"},
    {"name": "Company 3"}
]

# Memanggil fungsi dan mencetak hasilnya
result = get_companies_with_employees(employees, companies)
print(result)
````

### 1. ANALISA

````python
# Membaca dataset
df = pd.read_csv("50_Startups.csv")

# Menampilkan field mana saja yang memiliki data kosong
fields_with_missing_data = df.columns[df.isnull().any()]
print("Fields dengan data kosong:", fields_with_missing_data)

# Mengisi data kosong dengan nilai mean
for field in fields_with_missing_data:
    mean_value = df[field].mean()
    df[field].fillna(mean_value, inplace=True)

# Menyimpan hasil perubahan
df.to_csv("50_Startups.csv", index=False)

df.sample(5)
````

### 2. OneHotEncoder

````python
# Membaca dataset
df = pd.read_csv("50_Startups.csv")

# Melakukan OneHotEncoder ke field State
df = pd.get_dummies(df, columns=['State'])

# Menyimpan hasil perubahan
df.to_csv("50_Startups_onehot.csv", index=False)

df.sample(5)
````

### 3. FIELD

````python
# Membaca dataset
df = pd.read_csv("50_Startups.csv")

# Menghitung nilai Tax
df['Tax'] = (df['Profit'] + df['Marketing Spend'] + df['Administration']) * 0.05

# Menyimpan hasil perubahan
df.to_csv("50_Startups_tax.csv", index=False)

df.sample(5)
````

### 4. STANDAR SCALER FIELD

````python
from sklearn.preprocessing import StandardScaler

# Membaca dataset
df = pd.read_csv("50_Startups.csv")

# Menghapus kolom non-numerik (State)
df_numerical = df.drop(columns=['State'])

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Melakukan scaling ke field-field numerik
scaled_data = scaler.fit_transform(df_numerical)

# Membuat DataFrame baru dari data yang sudah di-scale
df_scaled = pd.DataFrame(scaled_data, columns=df_numerical.columns)

# Menambahkan kolom State yang telah dihapus sebelumnya
df_scaled['State'] = df['State']

# Menyimpan hasil perubahan
df_scaled.to_csv("50_Startups.csv", index=False)

df.sample(5)
````
